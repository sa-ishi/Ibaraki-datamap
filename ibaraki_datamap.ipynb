### 地図表示のためには緯度経度情報が必要
- 市町村の緯度経度がわかるデータ  Code for FUKUI
- https://github.com/code4fukui/localgovjp
- 日本の地方自治体一覧オープンデータ から　市町村緯度経度情報を得てmergeする。

```python
import pandas as pd

#市町村のデータ(令和3年2月1日現在)
# https://www.pref.ibaraki.jp/kikaku/joho/it/opendata/od-05/010700_20190205_local.html
#県内市町村における人口,商業,工業及び財政力などに関する基礎データ
df = pd.read_csv('data/010700_20210210_local-table.csv', thousands=',', encoding = 'cp932')

# 市町村の緯度経度情報の表をlatlonとして読み、茨城県の部分をdf_ibaとして取り出す。
latlon = pd.read_csv('data/localgovjp-utf8.csv', encoding="utf-8")
df_iba = latlon[latlon['pref'] == "茨城県"]
"""
茨城県の市町村の緯度経度情報をlatlon_ibaとし、
県内市町村における人口,商業,工業及び財政力などに関する基礎データdfにmergeする。
基礎データに緯度経度データが備わり、地図表示できるデータとする。
"""
latlon_iba = df_iba[['lat', 'lng','city']]
df_m = pd.merge(latlon_iba, df, left_on='city', right_on='項目（令和３年（2021年）２月１日現在）')
df_m
```

```python
#交通事故発生件数（令和2年）［件］の地図表示をする
import numpy as np
import folium 
traff = df_m[['lat', 'lng', 'city', '交通事故発生件数（令和2年）［件］',
              '人口（令和3年1月1日）［人］']].values
m = folium.Map(location=[36.327222, 140.090556], zoom_start=10)

for data in traff:
    jiko =  data[3]*10000/data[4]
    pop_memo =  '{} 人口1万人当たり交通事故 {:.1f}件'.format(data[2], jiko)
    folium.Circle(radius=np.sqrt(jiko)*500, location=[data[0],data[1]], tooltip=pop_memo,
                  fill_color='#EDA098', fill_opacity=0.3, show=True).add_to(m)  
m.save('map/iba1万人当たり交通事故.html')  
```

```python
df_m['1万人当たり交通事故［件］'] = df_m['交通事故発生件数（令和2年）［件］'] / df_m['人口（令和3年1月1日）［人］'] * 10000
traff_jiko = df_m[['lat', 'lng', 'city', '交通事故発生件数（令和2年）［件］',
      '人口（令和3年1月1日）［人］', '1万人当たり交通事故［件］']]
traff_jiko
```

```python
#'自動車保有台数（平成30年度末）［台］' の地図表示をする

import numpy as np
import folium 
auto = df_m[['lat', 'lng', 'city', '自動車保有台数（平成30年度末）［台］',
              '人口（令和3年1月1日）［人］']].values
m = folium.Map(location=[36.3, 140.3], zoom_start=10)

for data in auto:
    autom =  data[3]/data[4]
    pop_memo =  '{}:1人当たり自動車保有台数 {:.1f}台'.format(data[2], autom)
    folium.Circle(radius=np.sqrt(autom)*2500, location=[data[0],data[1]], tooltip=pop_memo,
                  fill_color='#77B8DA', fill_opacity=0.3, show=True).add_to(m)  
m.save('map/iba1人当たり自動車保有.html')  
```

```python
#'人口（令和3年1月1日）［人］' 65歳以上の割合が少ないほど濃い色で人口が大きいほど大きな円　人口は円の面積に比例
import numpy as np
import folium

jinko = df_m[['lat', 'lng', 'city', '人口（令和3年1月1日）［人］','人口密度（令和3年1月1日）［人／k㎡］',
              '65歳以上人口割合（令和3年1月1日）［％］']].copy()
jinko_list = jinko.values
m = folium.Map(location=[36.3, 140.2], zoom_start=10)

for data in jinko_list:
    pop_memo =  '{} {:.0f}千人 65歳以上 {}%'.format(data[2], data[3]/1000, data[5])
    folium.Circle(radius=(np.sqrt(data[3])*10), location=[data[0],data[1]], tooltip=pop_memo,
                  fill_color='#EDA098', fill_opacity=(50-data[5])/100, show=True).add_to(m)  
m.save('map/iba人口_65歳以上令和3年.html')   
```
