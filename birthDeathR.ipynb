```python
# 茨城県市町村の出生率・死亡率をグラフで表わす。
%matplotlib inline
import matplotlib.pyplot as plt
import japanize_matplotlib

import pandas as pd
df = pd.read_csv('data/010700_20210210_local-table.csv', thousands=',', encoding = 'cp932')

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1,1,1)
data1 = df['出生率（人口千対）（令和1年）［‰］\n(人口千対)']
data2 = df['死亡率（人口千対）（令和1年）［‰］']
ax.violinplot([data1, data2])
ax.set_xticks([1,2])
ax.set_xticklabels(['出生率', '死亡率'])
ax.set_title('茨城県市町村 人口千人あたりの出生率・死亡率[‰]（令和1年）', size=16 )
ax.grid(axis='y', linestyle='--') # y軸に点線を描く
plt.savefig("graph/茨城県市町村-出生・死亡率.png")
plt.show()
```
